```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
keyword= 'laptop'
with urllib.request.urlopen('https://bhinneka.com/jual?cari='+keyword) as page:
    html= page.read()
    print(html)
```

```python
soup = bs(html,"lxml")
print(type(soup))
print(soup.prettify())
```

```python
produk= soup.find_all("div", "o_wsale_product_grid_wrapper podition-relative h-100")
i = 1
for i in produk : 
    nama = i.find('h6', 'o_wsale_products_item_title mb-2').get_text().replace('\n','')
```

```python
nama_produk = []
harga_produk = []
cicilan_produk = []
produk = soup.find_all("div", "o_wsale_product_grid_wrapper position-relative h-100")

i = 1
for i in produk :
    nama = i.find('h6', 'o_wsale_products_item_title mb-2').get_text().replace('\n','')
    harga = i.find('div','o_wsale_product_sub d-flex justify-content-between align-items-end pb-1').get_text().replace('\n','')
    cicilan =i.find('span', 'bmd-installment').get_text().replace('\n','')
   
    nama_produk.append(nama)
    harga_produk.append(harga)
    cicilan_produk.append(cicilan)
    
data = {'nama':nama_produk, 'harga':harga_produk, 'cicilan':cicilan_produk}
df = pd.DataFrame(data, columns = ['nama','harga','cicilan'])
df
```
